```python
from notion_client import Client

def find_and_update_notion_page(notion: Client, database_id: str, keyword: str, new_properties: dict):
    """
    Find pages with a specific keyword and update their properties.

    Args:
        notion (Client): An instance of the Notion client.
        database_id (str): The ID of the database to search.
        keyword (str): The keyword to look for.
        new_properties (dict): The new properties to set on the pages.
    """
    # Search for pages with the keyword
    filter = {
        "or": [
            {
                "property": "Name",
                "text": {
                    "equals": keyword
                }
            },
            {
                "property": "Tags",
                "multi_select": {
                    "contains": keyword
                }
            }
        ]
    }

    response = notion.databases.query(
        database_id=database_id,
        filter=filter
    )

    # Update the found pages
    for page in response['results']:
        notion.pages.update(
            page_id=page['id'],
            properties=new_properties
        )

# Usage
notion = Client(auth="secret_58C8Buv5zKhxXbJ5ijVQROuWzWQopv9dacFjOcWENS0")
database_id = "your-database-id"
keyword = "test"

new_properties = {
    "Name": {
        "title": [
            {
                "text": {
                    "content": "New Name"
                }
            }
        ]
    },
    "Tags": {
        "multi_select": [
            {
                "name": "New Tag"
            }
        ]
    }
}

find_and_update_notion_page(notion, database_id, keyword, new_properties)
```
